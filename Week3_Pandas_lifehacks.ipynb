```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv('movie_metadata.csv')
```

```python
movie.head()
```

```python
genres = movie[['movie_title', 'genres']]
```

```python
def reversator(value):
    return value[::-1]
    
```

```python
for row in genres.values:
    for value in row:
        print(reversator(value))
```

```python
for i, row in genres.iterrows():
    print(row.map(reversator))
```

```python
budget = movie[['budget', 'duration']]
```

```python
%%time
for row in genres.values:
    for value in row:
        print(reversator(value), end='')
```

```python
%%time
#budget.applymap(lambda x: x+1)
budget.apply(np.mean, axis=0)
```

## Time Series

```python
stock = pd.read_csv('https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/amzn_stock.csv', index_col='Date', parse_dates=True)
```

```python
stock.head()
```

```python
stock.resample('1y').mean()
```

```python
stock.rolling(3, min_periods=1).mean()
```

```python
np.mean(stock.index.dayofyear)
```

```python
stock.index.day.value_counts()
```
