```python
def plot_2d_circles(data,colors):
    plt.figure(figsize=(8,8))
    plt.scatter(list(map(lambda x: x[0], data[0])), list(map(lambda x: x[1], data[0])), c=data[1], cmap=colors)
    plt.show()
```

```python
noisy_cirlces = datasets.make_circles(noise=0.1)
plot_2d_circles(noisy_cirlces, colors)
```

# Linear models in sklearn

```python
from sklearn import datasets, model_selection, linear_model, metrics
from matplotlib.colors import ListedColormap
```

```python
centers = datasets.make_blobs(centers=2, cluster_std=5.5, random_state=1)
```

```python
colors = ListedColormap(['red', 'blue'])
```

```python
plot_2d_circles(centers, colors)
```

```python
train_x, test_x, train_y, test_y = model_selection.train_test_split(centers[0], centers[1], \
                                                                    test_size=0.3, random_state=1)
```

```python
ridge = linear_model.RidgeClassifier()
```

```python
ridge.fit(train_x, train_y )
```

```python
predicted = ridge.predict(test_x)
```

```python
predicted
```

```python
metrics.accuracy_score(test_y, predicted)
```

```python
ridge.coef_
```

```python
ridge.intercept_
```

# Lodistic regression 

```python
log_regression = linear_model.LogisticRegression()
```

```python
log_regression.fit(train_x, train_y)
```

```python
log_regression.predict(test_x)
```

```python
ridge_class = model_selection.cross_val_score(ridge, centers[0], centers[1], scoring='accuracy', cv=10)
```

```python
lr_class = model_selection.cross_val_score(log_regression,centers[0], centers[1], scoring='accuracy', cv=10)
```

```python
print(f"Ridge mean:{np.mean(ridge_class)}, ridge max:{max(ridge_class)}, ridge min:{min(ridge_class)}")
print(f"Ridge mean:{np.mean(lr_class)}, ridge max:{max(lr_class)}, ridge min:{min(lr_class)}")
```

```python
scorer = metrics.make_scorer(metrics.accuracy_score)
```

```python
cv_strategy = model_selection.StratifiedShuffleSplit(n_splits=20, test_size=0.2, random_state=1)
```

```python
ridge_class = model_selection.cross_val_score(ridge, centers[0], centers[1], scoring=scorer, cv=10)
lr_class = model_selection.cross_val_score(log_regression,centers[0], centers[1], scoring=scorer, cv=10)

print(f"Ridge mean:{np.mean(ridge_class)}, ridge max:{max(ridge_class)}, ridge min:{min(ridge_class)}")
print(f"Ridge mean:{np.mean(lr_class)}, ridge max:{max(lr_class)}, ridge min:{min(lr_class)}")
```
