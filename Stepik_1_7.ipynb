```python
stud_perf = pd.read_csv('./StudentsPerformance.csv')
```

```python
stud_perf.columns = [x.replace(' ','_') for x in stud_perf.columns]
```

```python
stud_perf.math_score.hist()
```

```python
stud_perf.plot.scatter(x='math_score', y='reading_score')
```

```python
ax = sns.lmplot(x='math_score', y='reading_score', hue='gender', data=stud_perf, fit_reg=False)
ax.set_xlabels("Math score")
ax.set_ylabels('Reading score')
ax.title('My plot')
```
